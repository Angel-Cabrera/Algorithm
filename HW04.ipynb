```python
def recursive4(n,m,o):
    
    if(n<=0):
        print(m,o)
        y=n
        y=m
    else:
        recursive4(n-1, m+1, o)
        recursive4(n-1,m,o+1)
```

```python
def time_c(n,m,o):
    start = time.time()
    recursive4(m,n,o)
    end = time.time()
    com = end - start
    return com
```

```python
n= []
time_comp=[]
for i in range (1,10,1):
    time_temp = time_c(n=i, m=i, o=i)
    n.append(i)
    time_comp.append(time_temp)
```
